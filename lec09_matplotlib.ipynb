#### 선 그래프 그리기

- 제목과 x축, y축 레이블 설정
  - title(), xlabel(), ylabel() 함수


```python
import matplotlib.pyplot as plt

plt.plot(['kim', 'lee', 'kang'], [85, 88, 90])

plt.title('English Score of three students') # 그래프 제목 설정
plt.xlabel('Student Name') # x축 레이블 설정
plt.ylabel('Score') # y축 레이블 설정

plt.show()
```

- 한글 폰트 사용
  - rc 모듈의 rc() 함수

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic') # font로 맑은 고딕 폰트를 설정 - 여기서는 영문으로 사용해야 함

font1 = {'size':18, 'color':'green'} # 제목의 폰트 크기와 색상을 설정하기 위해 사용하는 딕셔너리

xdata = ['안지영', '홍지수', '황예린']
ydata = [90, 85, 88]
plt.plot(xdata, ydata)

plt.title('세명 학생의 국어 성적', fontdict=font1) # 한글 제목 사용, 폰트 크기, 색상 설정을 위해 fontdict 파라미터 이용
plt.xlabel('이름') # x축 레이블 한글 사용
plt.ylabel('성적') # y축 레이블 한글 사용

plt.show()
```

- 범례 (legend) 표시
  - legend() 함수

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88] # 국어 점수
ydata2 = [83, 88, 91] # 영어 점수
plt.plot(xdata, ydata1, label='국어') # 국어 점수를 나타내는 그래프의 범례(legend)의 레이블을 '국어'로 설정
plt.plot(xdata, ydata2, label='영어') # 영어 점수를 나타내는 그래프의 범례(legend)의 레이블을 '영어'로 설정
plt.legend(loc='upper center') # 범례(legend)를 화면에 표시, loc 파라미터 - 범례의 위치 설정
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

plt.title('세명 학생의 국어, 영어 성적')

plt.show()
```

- 선 스타일 설정
  - plot() 함수의 color, linestyle, marker 파라미터

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88] # 국어 점수
ydata2 = [83, 88, 91] # 영어 점수
ydata3 = [85, 97, 78] # 수학 점수
ydata4 = [92, 88, 82] # 사회 점수

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o') # '-' 실선, 'o' circle
plt.plot(xdata, ydata2, label='영어', color='#00ffff', linestyle='--', marker='x') # '--' 파선, 'x' x
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s') # '-.' 일점쇄선, 's' square
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d') # ':' 점선, 'd' diamond

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

- x, y 축 범위와 눈금 설정
  - xlim(), ylim(), xticks(), yticks() 함수
  - xticks(ticks=None, labels=None, **kwargs) https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.xticks.html
    - ticks: x축 눈금 위치를 나타내는 리스트 / labels: x축 눈금 레이블 (ticks 파라미터가 있는 경우만 가능)
  - grid() 함수

```python
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2 * x
    ydata.append(y)

plt.plot(xdata, ydata, label='y=2x')
plt.legend(loc='best')
plt.xlim(0, 35) # x축의 범위
plt.ylim(0, 70) # y축의 범위
plt.xticks(list(range(0, 35, 2))) # x축의 눈금
plt.yticks(list(range(0, 70, 5))) # y축의 눈금

plt.grid(True) # 그래프에 그리드 표시
#plt.grid(False) # 그리드 표시하지 않음
plt.show()
```

#### 다중 그래프 그리기

- subplots() 이용
  - https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.subplots.html
        

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = ['월', '화', '수', '목', '금', '토', '일']
y1 = [6.5, 5.7, 5.5, 6.7, 6.3, 7.5, 8.3]
y2 = [6.3, 7.7, 7.5, 7.7, 6.2, 7.3, 8.5]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9, 3), sharex=True, sharey=True)
# fig - Figure 객체
# axs - Axes 객체 배열. 이를 이용해서 서브 그래프를 그림 
# nrows - 행 개수
# ncols - 열 개수
# figsize - 그래프 크기 (가로, 세로)
# sharex - 같은 x 축 눈금 사용
# sharey - 같은 y 축 눈금 사용

ax = axs[0]
ax.scatter(x,y1) # 산포 그래프 그리기
ax.set_title('2018') # 서브 그래프의 제목 설정

ax = axs[1]
ax.scatter(x,y2)
ax.set_title('2019')

fig.suptitle('연간 요일별 평균 수면시간') # 전체 그래프 제목 설정

plt.show()
```

- 그래프 사이 간격 조정
  - subplots_adjust() 함수

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

x = list(range(1,11))
y = list(range(10, 101, 10))

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(9, 5), sharex=True, sharey=True)
ax = axs[0][0]
ax.plot(x,y)
ax.set_title('선 그래프 1')

ax = axs[0][1]
ax.plot(x,y, color='red', linestyle='--', marker='o')
ax.set_title('선 그래프 2')

ax = axs[1][0]
ax.bar(x,y)
ax.set_title('막대 그래프')

ax = axs[1][1]
ax.scatter(x,y)
ax.set_title('산포 그래프')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.5, hspace=0.5)
# left, right, bottom, top - 좌, 우, 하단, 상단 여백
# wspace - 가로 간격. 창 가로 크기의 비율 
# hspace - 세로 간격. 창 세로 크기의 비율

plt.show()
```
