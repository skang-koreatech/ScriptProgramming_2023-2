- DataFrame.groupby()
  - 데이터를 그룹별로 분할하여 각 그룹에 대해 데이터 처리를 하거나 그룹별 통계량을 확인하고자 할 때 유용
  - 통계 결과는 통계 계산이 가능한 수치형 컬럼에 대해서만 산출
  - count, sum, mean, median, var, std, min, max, prod (곱)
  - first, last

```python
weather.groupby('month')
```

```python
weather.groupby('month').count()
```

```python
weather.groupby('month').mean()
```

```python
weather.groupby('month').std()
```

```python
weather.groupby('month').agg(['first', 'last', 'mean', 'var'])
```

```python
# 월별 평균기온 그래프

means = weather.groupby('month').mean()
means['평균기온'].plot(kind='bar')
```

- 결측치 처리하기
  - 결측치, 누락값 (missing data/values): NaN, Na, Null, None 
  - isna(), isnull(): 결측치가 있는지 확인
  - fillna(): 결측치를 다른 값으로 교체
  - dropna(): 결측치가 하나라도 포함된 행 데이터를 제거

```python
weather = pd.read_csv('./weather.csv', index_col=0, encoding='CP949')
```

```python
weather[weather['최대풍속'].isna()]
```

```python
missing = weather[weather['평균풍속'].isna()]
missing
```

- pandas.DataFrame.dropna(axis=0, how='any', inplace=False)
  - axis=0: 결측치를 포함한 행 삭제
  - axis=1: 결측치를 포함한 열 삭제
  - how='any': 결측치가 하나라도 포함되면 제거
  - how='all': 행 또는 열 모두가 결측치여야 제거
  - inplace=True: 원본 데이터에서 결측치 삭제
  - inplace=False: 원본은 그대로 두고 결측치가 삭제된 데이터프레임 반환

```python
print(missing.dropna())
missing
```

```python
missing.fillna(0, inplace=True)
missing
```

#### DataFrame 결합

- pandas.concat(df_list, axis=0, join='outer')
  - df_list: 결합할 DateFrame의 리스트
  - axis: 0이면 테이블의 행을 늘려서 결합, 1이면 테이블의 열을 늘려서 결합
  - join: 테이블을 결합할 때 레이블을 어떻게 사용할지 결정. 'outer'이면 레이블의 합집합으로 생성하고 'inner'이면 레이블의 교집합으로 생성

```python
df_1 = pd.DataFrame( {'A' : ['a10', 'a11', 'a12'],
                      'B' : ['b10', 'b11', 'b12'],
                      'C' : ['c10', 'c11', 'c12']} , index = ['가', '나',  '다'] )

df_2 = pd.DataFrame( {'B' : ['b23', 'b24', 'b25'],
                      'C' : ['c23', 'c24', 'c25'],
                      'D' : ['d23', 'd24', 'd25']} , index = ['다', '라',  '마'] )
```

```python
df_1
```

```python
df_2
```

```python
df_3 = pd.concat([df_1, df_2])
df_3
```

```python
pd.concat([df_1, df_2], join='inner')
```

```python
pd.concat([df_1, df_2], axis=1, join='outer')
```

#### DataFrame 병합

- DataFrame.merge(right, how='inner', on=None)
  - right: 현재의 DateFrame과 결합할 DateFrame
  - how: 병합 방식 ('inner': 양쪽 테이블 모두에 존재하는 키 조합 사용, 'outer': 양쪽 테이블에 존재하는 모든 키 조합 사용, 'left': 왼쪽 테이블에 존재하는 모든 키 조합 사용, 'right': 오른쪽 테이블에 존재하는 모든 키 조합 사용)
  - on: 병합 기준 컬럼

```python
df_1
```

```python
df_2
```

```python
df_3 = df_1.merge(df_2, how='outer', on='B')
df_3
```

```python
df_4 = df_1.merge(df_2, how='right', on='B')
df_4
```

```python
df_5 = df_1.merge(df_2, how='outer', on='C')
df_5
```

```python
# 인덱스를 키로 이용하여 merge 하기

df_1.merge(df_2, how ='outer', left_index = True, right_index = True )
```

#### 함수 적용과 매핑

- DataFrame 객체에도 유니버설 함수(배열의 각 원소에 적용되는 메서드)를 적용할 수 있음

```python
import numpy as np

frame = pd.DataFrame(np.random.randn(4, 3), columns=list('abc'), 
                    index=['AAA', 'BBB', 'CCC', 'DDD'])
frame
```

```python
f = lambda x: x.max() - x.min()

frame.apply(f)
```

```python
frame.apply(f, axis='columns')
```

```python
def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])

frame.apply(f)
```

- DataFrame의 모든 개별 원소에 함수 매핑
  - applymap()

```python
frame
```

```python
def add_10(x):
    return x + 10

frame.applymap(add_10)
```

```python
frame.applymap(lambda x: abs(x))
```
